```python
import pandas as pd
```

```python
englishDF = pd.read_csv("/content/drive/MyDrive/data/english/englishData.csv",
                  on_bad_lines='skip',
                  encoding='ISO-8859-1'
                )
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Build Histogram
plt.hist(englishDF['CELF_SCORING'], color = 'yellow', edgecolor = 'black',
         bins = int(10), label = "CELF")

# Seaborn Histogram
sns.histplot(englishDF['TOLD_SCORING'], kde=False,
             bins = int(10), color = 'pink', label = "TOLD")

# Add Labels
plt.title('ENGLISH Dataset - Histogram of Scoring')
plt.xlabel('Scores')
plt.ylabel('Sentences')
plt.legend(loc="upper left")
```

```python
x1 = list(englishDF[englishDF['STRUCTURE']=='S_SVO+1_Aux']['CELF_SCORING'])
x2 = list(englishDF[englishDF['STRUCTURE']=='S_WH-quest.']['CELF_SCORING'])
x3 = list(englishDF[englishDF['STRUCTURE']=='S_Long_Pass.']['CELF_SCORING'])
x4 = list(englishDF[englishDF['STRUCTURE']=='S_Adjunct']['CELF_SCORING'])
x5 = list(englishDF[englishDF['STRUCTURE']=='S_Obj.Rel_RB']['CELF_SCORING'])
x6 = list(englishDF[englishDF['STRUCTURE']=='S_SVO+2_Aux']['CELF_SCORING'])
x7 = list(englishDF[englishDF['STRUCTURE']=='S_Short_Pass.']['CELF_SCORING'])
x8 = list(englishDF[englishDF['STRUCTURE']=='S_Cond.']['CELF_SCORING'])
x9 = list(englishDF[englishDF['STRUCTURE']=='S_Obj.Rel_CE']['CELF_SCORING'])


# Assign colors for each airline and the names
colors = ['#E69F00', '#56B4E9', '#F0E442', '#009E73', '#D55E00', '#F0E562', '#D25F00', '#003E99', '#E23F11']
names = ['SVO+1_Aux', 'WH-quest.', 'Long_Pass.', 'Adjunct',
       'Obj.Rel_RB', 'SVO+2_Aux', 'Short_Pass.', 'Conditional',
       'Obj.Rel_CE']

plt.hist([x1, x2, x3, x4, x5, x6, x7, x8, x9], bins = int(180/15),
         color = colors, label=names)

# Plot formatting
plt.legend()
plt.xlabel('Scores')
plt.ylabel('Sentences')
plt.title('ENGLISH Dataset - Sentence Structure Histogram')
```
